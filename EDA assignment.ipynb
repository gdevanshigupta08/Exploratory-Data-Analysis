```python
#Creating predefined function to plot pie chart
def uni_categorical_pie(dataframe,categorical_column):
    fig=plt.figure(figsize=(12,6))
    #plotting pie chart for Repayers
    dataframe[categorical_column].value_counts().plot.pie(autopct='%1.1f%%',textprops={'color': 'black',"fontsize":15})
    plt.title("Plotting pie chart for "+ categorical_column,fontdict={'fontsize' : 15, 'fontweight' :5, 'color' : 'Blue'})
    plt.show()
```

```python
#Creating predefined function to plot count and bar chart
def uni_categorical_bar(dataframe,categorical_column):
    fig,(ax1,ax2)=plt.subplots(1,2,figsize=(12,6))
    
    #Countplot on the basis of target
    sns.countplot(x=categorical_column,data=dataframe,ax=ax1)
    plt.title("Plotting countplot for "+ categorical_column,fontdict={'fontsize' : 15, 'fontweight' : 3, 'color' : 'Blue'})
    
    #Percentage of target with respect to the categorical_column
    df=pd.DataFrame()
    df['0']=((Repayer_target[categorical_column].value_counts())/len(Repayer_target))*100
    df['1']=((Defaulter_target[categorical_column].value_counts())/len(Defaulter_target))*100
    df.plot.bar(ax=ax2)
    plt.title("Plotting barchart in percentage for "+categorical_column,fontdict={'fontsize' : 15, 'fontweight' : 3, 'color' : 'Blue'})
    
    fig.tight_layout()
    ax1.tick_params(axis='x', rotation=90,size=12)
    ax2.tick_params(axis='x', rotation=90,size=12)   
    plt.show()
```

```python
#Creating predefined function to plot count and bar chart
def uni_categorical_barh(dataframe,categorical_column):
    fig,(ax1,ax2)=plt.subplots(nrows=2,figsize=(20,30))
    
    #Countplot on the basis of target
    sns.countplot(y=categorical_column,data=dataframe,ax=ax1)
    plt.title("Plotting countplot for "+ categorical_column,fontdict={'fontsize' : 20, 'fontweight' : 3, 'color' : 'Blue'})
    ax1.tick_params(axis='x', rotation=90,labelsize=20)
    ax1.tick_params(axis='y',labelsize=20)
    ax1.set_xlabel("COUNT",fontdict={'fontsize' : 20, 'fontweight' :20, 'color' : 'Blue'})
    ax1.set_ylabel(categorical_column,fontdict={'fontsize' : 20, 'fontweight' : 3, 'color' : 'Blue'})
    
    #Percentage of target with respect to the categorical_column
    df=pd.DataFrame()
    df['0']=((Repayer_target[categorical_column].value_counts())/len(Repayer_target))*100
    df['1']=((Defaulter_target[categorical_column].value_counts())/len(Defaulter_target))*100
    df.plot.barh(ax=ax2)
    ax2.tick_params(axis='x', rotation=90,labelsize=20)
    ax2.tick_params(axis='y',labelsize=20)
    ax2.set_xlabel("Percentage",fontdict={'fontsize' : 20, 'fontweight' :20, 'color' : 'Blue'})
    ax2.set_ylabel(categorical_column,fontdict={'fontsize' : 20, 'fontweight' : 3, 'color' : 'Blue'})
    
    plt.title("Plotting barchart in percentage for "+categorical_column,fontdict={'fontsize' : 20, 'fontweight' :5, 'color' : 'Blue'})
    fig.tight_layout()
    plt.show()
    
```

```python
#Creating predefined function to plot countplot
def uni_categorical_bar2(dataframe,categorical_column):
    fig=plt.figure(figsize=(12,6))
    
    #Countplot on the basis of target
    sns.countplot(x=categorical_column,data=dataframe)
    plt.tick_params(axis='x', rotation=90)
    plt.title("Plotting countplot for "+ categorical_column,fontdict={'fontsize' : 15, 'fontweight' : 3, 'color' : 'Blue'})
```

### Numerical Univariate Analysis

```python
#Creating predefined function fro countplot
def uni_numerical_bar(dataframe,numerical_column):
    fig=plt.figure(figsize=(12,6))
    
    #Countplot on the basis of target
    sns.countplot(x=numerical_column,data=dataframe)
    plt.title("Plotting countplot for "+ numerical_column,fontdict={'fontsize' : 15, 'fontweight' : 3, 'color' : 'Blue'})
    
    fig.tight_layout()
    plt.xticks(rotation=90)  
    plt.show()
```

```python
def plot_percentage_of_defaulters(df, income_range_col, target_col):
    fig=plt.figure(figsize=(12,6))
    
    # Calculate the total count of each income range and the number of defaulters in each range
    income_range_counts = df[income_range_col].value_counts()
    defaulters_counts = df.groupby(income_range_col)[target_col].sum()

    # Calculate the percentage of defaulters for each income range
    per_defaulters = (defaulters_counts / income_range_counts) * 100
    data= pd.DataFrame({ income_range_col: per_defaulters.index,'Percentage_Defaulters': per_defaulters.values})
    
    data.sort_values(by='Percentage_Defaulters',ascending=False,inplace=True)
    sns.barplot(x=income_range_col, y='Percentage_Defaulters', data=data)

    plt.xlabel(income_range_col.capitalize())
    plt.ylabel('Percentage of Defaulters')
    plt.title('Percentage of Defaulters in Each ' + income_range_col.capitalize())

   
    plt.xticks(rotation=45)
    plt.show()
```

```python
def biv_analysis(value_col,index_col,col):
    plt.figure(figsize=(30,20))
    merged_dataframes.pivot_table(values=value_col,index=index_col,columns=col,aggfunc='mean').plot.bar()
   
    plt.xlabel(index_col,fontdict={'fontsize' : 20, 'fontweight' :20, 'color' : 'Blue'})
    plt.ylabel('Defaulters',fontdict={'fontsize' : 20, 'fontweight' :20, 'color' : 'Blue'})
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    
    plt.title(index_col +" vs "+ col,fontdict={'fontsize' : 15, 'fontweight' :20, 'color' : 'Blue'})
    plt.show()
```
